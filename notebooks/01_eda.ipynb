## Exploratory Data Analysis

In this notebook, we explore the dataset using statistical summaries and visualizations. The goal is to understand the data's structure, distributions, and relationships relevant to loan approval prediction.


## üîç Basic Data Audit

We inspect the dataset's size, column types, and completeness. This helps identify data quality issues such as missing values and formatting inconsistencies.


```python
# Import Necessary Libraries and Load The Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/train.csv")
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

## üîó Bivariate Analysis

We explore how features relate to the target variable (`loan_status`) to detect potential predictors.


### Categorical Features vs. Target



```python
# Person Home Ownership 
sns.countplot(data=df, x='person_home_ownership', hue='loan_status')
plt.title('Person Home Ownership Countplot by Loan Status')
plt.show()
print("Loan approval rate by person home ownership:")
print(pd.crosstab(df['person_home_ownership'], df['loan_status'], normalize='index') * 100)
print("\n")
```

## üìà Group Statistics

We group by the target variable and compute aggregated statistics (mean, median) across numerical features to identify patterns by class.


```python
# Group Statistics
numerical_cols = [
    'person_age',
    'person_income',
    'person_emp_length',
    'loan_amnt',
    'loan_int_rate',
    'loan_percent_income',
    'cb_person_cred_hist_length'
]

df.groupby('loan_status')[numerical_cols].agg(['mean', 'median'])
```

## üßÆ Correlation Statistics

We compute the Pearson correlation matrix among numerical variables to identify multicollinearity and potential feature redundancy.


```python
# Correlation Statistics

numerical_cols = [
    'person_age',
    'person_income',
    'person_emp_length',
    'loan_amnt',
    'loan_int_rate',
    'loan_percent_income',
    'cb_person_cred_hist_length',
    'loan_status'
]

corr_matrix = df[numerical_cols].corr(method='pearson')
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,fmt=".2f",linewidths=0.5, mask=np.triu(corr_matrix))
plt.title('Correlation Matrix')
plt.show()
```

## Interpretation and Insights

This exploratory analysis revealed several key patterns relevant for predicting loan approvals, including:

- **Home Ownership**: Renters faced the highest rejection rate (**22.3%**), while owners had the lowest (**1.4%**).
- **Loan Grade**: Rejection rates rise steeply from grade A (**5%**) to grade G (**82%**).
- **Default History**: Applicants with prior defaults were rejected **30%** of the time, compared to **11.5%** without defaults.
- **Numerical Trends**: Rejected applicants tended to have lower income, shorter employment and credit history, and higher interest rates and loan-to-income ratios.

These insights provide a strong foundation for building our predictive model.
